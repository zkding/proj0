Load data as pandas dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

df = pd.read_csv('./data/screening_exercise_orders_v201810.csv')
```

View data

```python
df
```

add new column to show week info

```python
df['date'] = pd.to_datetime(df['date'])
df['Week'] = df['date'].apply(lambda x: "%d" % x.week)
df
```

noticed first date is Sunday, data need to be modified

```python
for i, row in df.iterrows():
    if row['date']<pd.Timestamp('2017-01-02 00:00:00'):
        df.at[i,'Week']='0'
df
```

assemble data to plot

```python
plotdata = df.groupby(['Week']).size()
dfplot = pd.DataFrame(plotdata)
dfplot.index = dfplot.index.map(int)
dfplot=dfplot.sort_index()
```

```python
dfplot.plot(figsize=(12,5))
```
